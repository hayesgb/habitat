```python
import pandas as pd
```

```python
import filenames
```

```python
giving = pd.read_csv(filenames.giving, index_col='id', low_memory=False)
personal = pd.read_csv(filenames.personal, index_col='id', low_memory=False)
```

```python
df = pd.merge(giving, personal, left_index=True, right_index=True)
```

```python
df.dtypes
```

```python
dates = ['first appeal date', 'first gift date', 'second gift date']
for d_ in dates:
    df[d_] = pd.to_datetime(df[d_], yearfirst=True)
```
